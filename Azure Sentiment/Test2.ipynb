```python
df = spark.read.format("csv").option("header","true").load("Files/ds_test.csv")
# df now is a Spark DataFrame containing CSV data from "Files/ds_test.csv".
display(df)
```

```python
import synapse.ml.core
from synapse.ml.services import AnalyzeText
from pyspark.sql.functions import col
from pyspark.sql.functions import col, to_date
from pyspark.sql.utils import AnalysisException
```

```python
mdl = (AnalyzeText()
        .setTextCol("phrase")
        .setKind("SentimentAnalysis")
        .setOutputCol("response")
        .setErrorCol("error"))
```

```python
outcome = mdl.transform(df)
```

```python
sentimentdataframe = outcome.withColumn("predicted_sentiment", col("response.documents.sentiment"))
```

```python
finalsentiment = sentimentdataframe.drop("error","response")
```

```python
display(finalsentiment)
```

```python
from pyspark.sql.functions import col
from pyspark.ml.evaluation import MulticlassClassificationEvaluator




accuracy_df = finalsentiment.withColumn("correct", 
                                      (col("predicted_sentiment") == col("sentiment")).cast("int"))

accuracy = accuracy_df.select("correct").agg({"correct": "avg"}).collect()[0][0]
print(f"Accuracy: {accuracy:.2f}")
```
